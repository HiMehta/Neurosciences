>**Description** :
The below Model of simple arm exhibits reachability capabilities of simple arm which generates reachable region depending on  *Humerus bone (upper arm)* length and *Ulna bone (lower arm)* length and generates elbow and shoulder joints and actual linear space of targets . Further this experiments distributed coding on taking targets from reachable regions and analyzing errors between actual position and position after trying to obtain mapping from normally distributed normalized values.

```python
import numpy as np
#prec = np.float64
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
n = 10
# Humerus bone (upper arm) length
l1 = 1.0

# Ulna bone (lower arm) length
l2 = 1.0

# Shoulder joint
Sx = 0.0
Sy = 0.0

#Domain of reachable Targets
domain = np.zeros((2,2))
domain[0] = [-0.5,0.25]
domain[1] = [1.0,1.75]
xmin,xmax = domain[0]
ymin,ymax = domain[1]

#print xmin,xmax
#weights
Theta = np.zeros((2,n*n))
#Error
error = np.zeros((n,n))
new_pos = np.zeros((2,n*n))
#No of Muscles
nm = 2
S = 0.07
```

```python
#Forward and inverse Definitions
def forward(theta1, theta2, l1=l1, l2=l2, Sx=Sx, Sy=Sy):
    # Elbow joint
    Ex = Sx + l1 * np.cos(theta1)
    Ey = Sy + l1 * np.sin(theta1)

    # Hand Position
    Hx = Ex + l2 * np.cos(theta1+theta2)
    Hy = Ey + l2 * np.sin(theta1+theta2)

    return Hx, Hy

def coordinations(theta1, theta2, l1 = 1., l2 = 1., pos0 = np.array([0.0, 0.0]).reshape(1,2)):
    x1 = np.cos(theta1) * l1 + pos0[0][0]
    y1 = np.sin(theta1) * l1 + pos0[0][1]
    x2 = np.cos(theta1 + theta2) * l2 + x1
    y2 = np.sin(theta1 + theta2) * l2 + y1
  
    return np.array([x2, y2]).reshape(1,2)


def inverse(Hx, Hy, l1=l1, l2=l2, Sx=Sx, Sy=Sy):
    Hx = np.array(Hx) - Sx
    Hy = np.array(Hy) - Sy

    # Elbow joint
    c2 = (Hx*Hx + Hy*Hy - l1*l1 - l2*l2)/(2*l1*l2)
    s2 = np.sqrt(1. - c2*c2)
    theta2 = np.arctan2(s2, c2)

    # Shoulder joint
    k1 = l1 + l2*c2
    k2 = l2*s2
    theta1 = np.arctan2(Hy, Hx) - np.arctan2(k2, k1)

    return theta1, theta2

def muscles(theta1, theta2):
    g1, g3 = 0., 0.
    g2 = g1 - theta1/np.pi
    g4 = g3 - theta2/np.pi
    #print g1,g2,g3,g4
    g = np.array([g1, g2, g3, g4]).reshape((nm,1))
    return g
```

```python
#Reachable Region
T1,T2 = np.meshgrid(np.linspace(0, 2*np.pi/3, n),
                    np.linspace(0, 2*np.pi/3, n))
Hx, Hy = forward(T1, T2)
f = plt.figure(figsize=(11,8))
ax = f.add_subplot(111,aspect=1)
ax.scatter(Hx, Hy, s=15,c="black")
plt.title("Reachability of Arm")
```

```python
def Gaussian(x,y,sigma=0.2,shape=(10,10)):
    X = np.linspace(0,1,shape[0],endpoint=True)
    Y = np.linspace(0,1,shape[1],endpoint=True)
    X,Y = np.meshgrid(X,Y)
    R = np.sqrt((X-x)**2 + (Y-y)**2)
    return np.exp(-R**2/sigma**2)
#test
if 0:
    G = Gaussian(0.5,0.5,shape=(50,50),sigma=0.1)
    plt.imshow(G, cmap=plt.cm.gray_r,interpolation="nearest",extent=[0,1,0,1])
```

```python
#Limites the vector to the proper min and max
def limits(W, Wmin = -0.5, Wmax = 0.5):
    
    N = np.random.normal(0.0, 0.005, W.shape)
    N = np.minimum(np.maximum(N, 0.0),1.0)
    W = W*(Wmin + (Wmax - Wmin)*N)
    return W
#Weights initialization
def initial_weights():
    Wmin, Wmax = -0.5, 0.5 
    #W = np.zeros((nm,n*n),dtype=np.float64)
    W = limits(np.random.normal(0.0,0.5,(nm,n*n)))
    return W
```

```python
def Encode(x,y,sigma,domain, shape=(10,10)):
    xmin,xmax = domain[0]
    ymin,ymax = domain[1]
    xm = (x-xmin) / (xmax-xmin)
    ym = (y-ymin) / (ymax-ymin)
    return Gaussian(xm,ym,sigma=sigma,shape=shape)
#test
if 0:
    SMA = Encode(xmin+(xmax-xmin)/2, ymin+(ymax-ymin)/2, sigma=0.25, domain=domain)
    #plt.imshow(SMA)
```

###### Pre computations of weights

```python
Theta = np.zeros((2,n*n))
new_pos = np.zeros((2,100))
error = np.zeros((10,10))
f = plt.figure(figsize=(10,10))
errorp = np.zeros((10,10))
xi = np.linspace(xmin,xmax, n)
yi = np.linspace(ymin,ymax, n)
X,Y = np.meshgrid(xi,yi)
target = np.zeros((1,2))
T1,T2 = inverse(X,Y) #Getting all Angles for above square
Theta[0] = T1.reshape(n*n)
Theta[1] = T2.reshape(n*n)
xnew = np.linspace(xmin,xmax, 10)
ynew = np.linspace(ymin,ymax, 10)
for i,x in enumerate(xnew):
    for j,y in enumerate(ynew):
        SMA = Encode(x,y,sigma=S,domain=domain,shape=(n,n))
        x_, y_ = np.dot(Theta,SMA.reshape((n*n,1))) / np.sum(SMA)
        pos = forward(x_,y_)
        new_pos[0][i*n+j] = pos[0]
        new_pos[1][i*n+j] = pos[1]
        error[i][j] = np.sqrt((x-pos[0])**2 + (y-pos[1])**2)
        errorp[i][j] = error[i][j]
        errorp[i][j] /= (xmax-xmin)/100
f = plt.figure(figsize=(15,11))
ax = f.add_subplot(121,aspect=1)
plt.title("Errors(Pre Computed Weights)")
im = plt.imshow(error, interpolation="nearest",cmap=plt.cm.gray_r,extent=[xmin,xmax,ymin,ymax])
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(im, cax=cax)
ax = f.add_subplot(122,aspect=1)
im = plt.imshow(errorp, interpolation="nearest",cmap=plt.cm.gray_r,extent=[xmin,xmax,ymin,ymax])
divider = make_axes_locatable(ax)
plt.title("%Errors(Pre Computed Weights)")
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(im, cax=cax)
plt.show()
```

###### Learning simulation

```python
W = initial_weights()
S = 0.08
n=10
h_MC = 0.09
epochs  = 200000
positions = np.zeros((n*n,2))
D = np.zeros(epochs)
xi = np.linspace(xmin,xmax, n)
yi = np.linspace(ymin,ymax, n)
#X,Y = np.meshgrid(xi,yi)
index = 0
for valx in xi:
    for valy in yi:
        positions[index][0] = valx
        positions[index][1] = valy
        index = index + 1
I = np.random.random_integers(0,n*n-1,epochs)
for k in range(epochs):
    pos = positions[I[k]]
    x = np.random.uniform(xmin,xmax)
    y = np.random.uniform(ymin,ymax) 
    theta1, theta2 = inverse(x,y)
    SMA = Encode(x,y,sigma=S,domain=domain,shape=(n,n))
    Motor = np.dot(W,SMA.reshape((n*n,1))) / np.sum(SMA)
    W = W + np.dot((np.array([theta1,theta2]).reshape((2,1)) - Motor),SMA.reshape((n*n,1)).T) * h_MC
    Motor = np.dot(W,SMA.reshape((n*n,1))) / np.sum(SMA)
    Arm = Motor
    cur_pos = forward(Arm[0], Arm[1])
    D[k] = np.sqrt((x-cur_pos[0])**2 + (y-cur_pos[1])**2)
f = plt.figure(figsize=(15,9))
ax = f.add_subplot(111)
ax.plot(D)
plt.title("Learning Errors in each epoch")
plt.show()
```
