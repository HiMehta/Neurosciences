#### Importing Packages and intializations.

```python
from dana import *
from decimal import *
import os
import csv
figs = {}
```

#### Explicit Definitions :-

```python
def draw_result(name, title=''):
    ''' '''
    axes = plt.gca()
    axes.grid()
    if(name == 'selection'):
        line, = axes.plot([],[],'r*')
        line1, = axes.plot([],[],'co')
        line3, = axes.plot([],[],'bs')
        plt.legend([line,line1,line3],['No selection', 'Any of two','Both'],'upper left',prop={'size':9})
    elif(name == 'channel Switching'):
        line, = axes.plot([],[],'r*')
        line1, = axes.plot([],[],'go')
        line2, = axes.plot([],[],'yo')
        line3, = axes.plot([],[],'bs')
        plt.legend([line,line1,line2,line3],['No selection', 'ch1','ch2','Both'],'upper left',prop={'size':9})
    axes.set_xlim([0.1,1.1])
    axes.set_ylim([0.1,1.2])
    axes.set_title(title, fontsize=12)
    for label in axes.get_xticklabels() + axes.get_yticklabels():
        label.set_fontsize(10)
    plt.xlabel('Channel 1', fontsize=12)
    plt.ylabel('Channel 2', fontsize=12)
    figs[name] = axes,line

def update_result(name, x, y,result):
    axes, line = figs[name]
    if(name == 'selection'):
        j=0
        markers = ['r*','co','co','bs']
        for k in markers:   
            tempx = [x[i] for i, m in enumerate(result) if m == j]
            tempy = [y[i] for i, m in enumerate(result) if m == j]
            axes.plot(tempx,tempy,k,ms=8)
            j=j+1
            plt.draw()
    elif(name == 'channel Switching'):
        j=0
        markers = ['r*','go','yo','bs']
        for k in markers:   
            tempx = [x[i] for i, m in enumerate(result) if m == j]
            tempy = [y[i] for i, m in enumerate(result) if m == j]
            axes.plot(tempx,tempy,k,ms=8)
            j=j+1
            plt.draw()
    plt.show()

#Reset Channels everytime ( No learning)

def Reset_all(St1_c,St2_c,STN_c,GPi_c,GPe_c,VLT_c,PFC_c,TRN_c):
    for i in range(6):
        St1_c[i]=St2_c[i]=STN_c[i]=GPi_c[i]=GPe_c[i]=VLT_c[i]=PFC_c[i]=TRN_c[i]=0.0   
    return(St1_c,St2_c,STN_c,GPi_c,GPe_c,VLT_c,PFC_c,TRN_c)

#adding data to file making ready to plot it later

def add_data_to_file(filename,content):
    output = open(filename, 'a')
    writer = csv.writer(output)
    writer.writerow(content)
    output.close()

#file Check delete if it exist so everytime it get filled with fresh data.   
def check_file(filename):
    if os.path.exists(filename):
        os.remove(filename)

def my_range(start, end, step):
    while start < end:
        yield start
        start += step

#Creating Range for plotting
def create_Range(xstart,xstop,xstep,ystart,ystop,ystep):
    x=[]
    y=[]
    for i in my_range(xstart,xstop,xstep):
        for j in my_range(ystart,ystop,ystep):
            x.append(round(i,3))
            y.append(round(j,3))
    return x,y

#Plotting from any two files (one contains range other contains data)
def plot_data(xdata,ydata,datafile,name):
    with open(datafile,'r') as d:
        reader1 = csv.reader(d)
        data = []
        for row in reader1:
            decision = map(int,row[0])
            data=data+decision
        #print len(xdata)
        data = data[:(len(xdata))]
        update_result(name,xdata,ydata,data)
```

#### Setting up necessary variables being used in groups

```python
#files
datafile = 'Gurney-no-learning-data.dat'
#rangefile = 'Gurney-plotting-range1.dat'
# Number of channels
n = 6

# Time constant
tau  =  0.04

# Resting levels
eSt1 = +0.20
eSt2 = +0.20
eSTN = -0.25
eGPe = -0.20
eGPi = -0.20
eVLT = +0.00
ePFC = +0.00
eTRN = +0.00

# Connections strength
D1 = 0.2
D2 = 0.2

STN_GPe = +0.9
St2_GPe = -1.0

STN_GPi = +0.9
GPe_GPi = -0.3
St1_GPi = -1.0

GPe_STN = -1.0
PFC_STN = +0.5
PC_STN  = +0.5

PFC_St1 = +0.5*(1+D1)
PC_St1  = +0.5*(1+D1)

PFC_St2 = +0.5*(1-D2)
PC_St2  = +0.5*(1-D2)

GPi_VLT = -1.0
PFC_VLT = +1.0
TRN_VLT = -0.4
TRN_VLT_self = -0.125

PFC_TRN = +1.0
GPi_TRN = -0.2
VLT_TRN = +1.0

VLT_PFC = +1.0

#plotting range and update 
start_x = 0.2
start_y = 0.2
stop_x = 1
stop_y = 1
Update_x_y = 0.1
duration = ((stop_x+Update_x_y)-start_x) / Update_x_y
```

#### Creating Dana Groups.

```python
# Posterior Cortex
PC  = zeros(n, '''V''')

# Striatum D1: medium spiny neurons of the striatum with D1 dopamine receptors
St1 = zeros(n, '''dU/dt = 1/tau*(-U + PC_ + PFC_)
                   V    = np.minimum(np.maximum(U-eSt1,0),1)
                   PC_; PFC_''')

# Striatum D2: medium spiny neurons of the striatum with D2 dopamine receptors
St2 = zeros(n, '''dU/dt = 1/tau*(-U + PC_ + PFC_)
                   V    = np.minimum(np.maximum(U-eSt2,0),1)
                   PC_; PFC_''')

# Sub-Thalamic Nucleus 
STN = zeros(n, '''dU/dt = 1/tau*(-U + PC_ + PFC_ + GPe_)
                   V    = np.minimum(np.maximum(U-eSTN,0),1)
                   PC_; PFC_; GPe_''')

# External Globus Pallidus
GPe = zeros(n, '''dU/dt = 1/tau*(-U + STN_ + St2_)
                   V    = np.minimum(np.maximum(U-eGPe,0),1)
                   STN_; St2_''')

# External Globus Pallidus
GPi = zeros(n, '''dU/dt = 1/tau*(-U + STN_ + St1_ + GPe_)
                   V    = np.minimum(np.maximum(U-eGPi,0),1)
                   STN_; St1_; GPe_''')

#  Ventro-Lateral Thalamus
VLT = zeros(n, '''dU/dt = 1/tau*(-U + PFC_ + TRN_ + GPi_)
                   V    = np.minimum(np.maximum(U-eVLT,0),1)
                   PFC_; TRN_; GPi_''')

# Prefrontal Cortex
PFC = zeros(n, '''dU/dt = 1/tau*(-U + PC_ + VLT_)
                   V    = np.minimum(np.maximum(U-ePFC,0),1)
                   PC_; VLT_''')

# Thalamic Reticular Nucleus
TRN = zeros(n, '''dU/dt = 1/tau*(-U + PFC_ + VLT_ + GPi_)
                   V    = np.minimum(np.maximum(U-eTRN,0),1)
                   PFC_; VLT_; GPi_''')
```

#### Creating Dana connections.

```python
# St1 connections
SparseConnection( PC('V'),  St1('PC_'),   PC_St1 * np.ones(1) )
SparseConnection( PFC('V'), St1('PFC_'), PFC_St1 * np.ones(1) )

# St2 connections
SparseConnection( PC('V'),  St2('PC_'),   PC_St2 * np.ones(1) )
SparseConnection( PFC('V'), St2('PFC_'), PFC_St2 * np.ones(1) )

# STN connections
SparseConnection( PC('V') , STN('PC_'),   PC_STN * np.ones(1) )
SparseConnection( PFC('V'), STN('PFC_'), PFC_STN * np.ones(1) )
SparseConnection( GPe('V'), STN('GPe_'), GPe_STN * np.ones(1) )

# GPe connections
DenseConnection(  STN('V'), GPe('STN_'), STN_GPe * np.ones((n,n)) )
SparseConnection( St2('V'), GPe('St2_'), St2_GPe * np.ones(1)     )

# GPi connections
DenseConnection(  STN('V'), GPi('STN_'), STN_GPi * np.ones((n,n)) )
SparseConnection( St1('V'), GPi('St1_'), St1_GPi * np.ones(1)     ) 
SparseConnection( GPe('V'), GPi('GPe_'), GPe_GPi * np.ones(1)     ) 

# VLT connections
SparseConnection( PFC('V'), VLT('PFC_'), PFC_VLT * np.ones(1) )
SparseConnection( GPi('V'), VLT('GPi_'), GPi_VLT * np.ones(1) ) 
K = TRN_VLT * np.ones((n,n))
K[np.arange(n),np.arange(n)] = TRN_VLT_self
DenseConnection(  TRN('V'), VLT('TRN_'), K )

# PFC connections
SparseConnection( VLT('V'), PFC('VLT_'), VLT_PFC * np.ones(1) )
SparseConnection(  PC('V'), PFC('PC_'),            np.ones(1) )

# TRN connections
SparseConnection( PFC('V'), TRN('PFC_'), PFC_TRN * np.ones(1) ) 
SparseConnection( VLT('V'), TRN('VLT_'), VLT_TRN * np.ones(1) ) 
SparseConnection( GPi('V'), TRN('GPi_'), GPi_TRN * np.ones(1) )
```

#### Giving Energy to PC of channel 1 and 2 after every 2 seconds and giving 2 seconds to stabilize .

```python
PC['V'][0] = start_x
PC['V'][1] = start_y
PC['V'][2] = 0.2
PC['V'][3] = 0.2
PC['V'][4] = 0.2
PC['V'][5] = 0.2
check_file(datafile)
#check_file(rangefile)
@clock.every(2*second)
def update_PC(t):
    if ((VLT['V'][0] == 0) and (VLT['V'][1] == 0)):
        is_select =0
    elif((VLT['V'][0] != 0) and (VLT['V'][1] != 0)):
        is_select =3
    elif((VLT['V'][0] != 0) and (VLT['V'][1] == 0)):
        is_select = 1
    else:
        is_select = 2
    add_data_to_file(datafile,[is_select])
   # add_data_to_file(rangefile, [round(PC['V'][0], 3),round(PC['V'][1], 3)])
    Reset_all(St1,St2,STN,GPi,GPe,VLT,PFC,TRN)
    PC['V'][1] = PC['V'][1]+Update_x_y
   # print VLT['V']
    VLT['V'].tolist()
    if (PC['V'][1] > stop_y):
        if (PC['V'][0]< stop_x):
            PC['V'][1] = start_y
            PC['V'][0] = PC['V'][0]+Update_x_y
    
@clock.at(duration*duration*2*second)
def plot(t):
    plt.ion()
    fig = plt.figure(figsize=(15,12))
    fig.patch.set_alpha(0.0)
    fig.suptitle('No Learning D1=D2=0.4', fontsize=12)
    plt.subplot(1,2,1,aspect=1)
    draw_result('selection','channel selection')
    plt.subplot(1,2,2,aspect=1)
    draw_result('channel Switching','channel Switching')
    x,y = create_Range(start_x,stop_x,Update_x_y,start_y,stop_y,Update_x_y)
    plot_data(x,y,datafile,'selection')
    plot_data(x,y,datafile,'channel Switching')
    plt.ioff()
```

#### Running the model for 10 seconds 

```python
# Run simulation for 10 second
run(time=duration*duration*2*second, dt=1*millisecond)
plt.ioff()
plt.savefig('D1D24GL.png',dpi=100)
```
